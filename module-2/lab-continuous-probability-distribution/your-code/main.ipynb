## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
x = uniform.rvs(size=10)
a = 2
b = 3
randoms  = a + (b-a)*x
print(randoms)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
#Funci√≥n

def uniform_dist(bottom=0,ceiling=0, count=0):
    return [random.uniform(bottom,ceiling) for i in range(count)]
```

```python
#Comprobaciones

print(len(uniform_dist(bottom=10, ceiling=15, count=100)))

uniform_dist(bottom=10, ceiling=60, count=1000)[:5]
```

```python
#Representaci√≥n Gr√°fica

fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)

x1 = uniform_dist(bottom=10, ceiling=15, count=100)
x2 = uniform_dist(bottom=10, ceiling=60, count=1000)

axs[0].hist(x1, bins=10)
axs[1].hist(x2, bins=10)

plt.show
```

How are the two distributions different?

Una de las principales caracter√≠sticas de las distribuciones uniformes corresponde al caso de que una variable aleatoria s√≥lo puede tomar valores comprendidos entre dos extremos "bottom" y "ceiling".

As√≠ en la primera gr√°fica, la probabilidad se distribuye entre el rango asignado de 10 y 15.

Por otro lado, en la segunda gr√°fica se distribuye entre 10 y 60. Adem√°s, por el mayor n√∫mero de muestras aportado, las probabilidades aumentan considerablemente para cada punto de dicho rango.





## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
#  Funci√≥n

def normal_dist(average,std,count):
    return [np.random.normal(average,std) for i in range(count)]
```

```python
#Comprobaciones

print(len(normal_dist(10,1,1000)))

normal_dist(10,1,1000)[:5]
```

```python
#Representaci√≥n Gr√°fica

fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)

x1 = normal_dist(10,1,1000)
x2 = normal_dist(10,50,1000)

axs[0].hist(x1, bins=50)
axs[1].hist(x2, bins=50)

plt.show
```

How are the two distributions different?


La media de ambas gr√°ficas se sit√∫an en 10, pero el gran n√∫mero de muestras y desviaci√≥n t√≠pica de la segunda gr√°fica provocan que dicha media quede "desvirtuada" al dibujarse.

Continuando por esta cuesti√≥n, la amplia desviaci√≥n t√≠pica de la segunda gr√°fica provoca una mayor dispersi√≥n de los datos en torno a la media respecto a la primera gr√°fica.

Por otro lado, la segunda gr√°fica presenta una mayor precisi√≥n en su representaci√≥n al disponer de mayor n√∫mero de muestras.
